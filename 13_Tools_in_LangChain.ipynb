# **Custom Tools**

```python
from langchain_core.tools import tool
```

```python
#step1: create a function
def multiply(a,b):
  """Multiply two numbers"""   #Highly recommended to add doc string - LLM understands
  return a*b
```

```python
#step2: Add type hints:
def multiply(a:int,b:int)->int:
  """Multiply two numbers"""
  return a*b
```

```python
#step3: add tool decorator:

@tool
def multiply(a:int,b:int)->int:
  """Multiply two numbers"""
  return a*b
```

```python
result = multiply.invoke({'a':4,'b':5})   #Hers, multiply is a tool with which LLM can interact
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())   #LLM sees this
```

# Method2 - Using Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):  #Pydantic class
  a:int = Field(required=True, description='The first number to add')
  b:int = Field(required=True, description='The second number to add')
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name='multiply',
    description='Multiply two numbers',
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':4,'b':5})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

# Method3- Using BaseTool class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):  #Pydantic class
  a:int = Field(required=True, description='The first number to add')
  b:int = Field(required=True, description='The second number to add')
```

```python
class MultiplyTool(BaseTool):
  name:str ='multiply'
  description:str ='Multiply two numbers'

  args_schema:Type[BaseModel]=MultiplyInput   #Type does type hinting

  def _run(self,a:int,b:int)->int:  #this func. name should be _run only
    return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```

# ToolKit

```python
from langchain_core.tools import tool

#custom tools:
@tool
def add(a:int,b:int)->int:
  """Add two numbers"""
  return a+b

@tool
def multiply(a:int,b:int)->int:
  """Multiply two numbers"""
  return a*b
```

```python
class MathToolkit:
  def get_tools(self):
    return [add,multiply]
```

```python
toolkit=MathToolkit()
tools=toolkit.get_tools()

print(tools)
```

```python
for tool in tools:
    print(tool.name, "=>", tool.description)
```

# **Currency Conversion Tool**

```python
# tool create
from langchain_core.tools import InjectedToolArg
from typing import Annotated

@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
  """
  This function fetches the currency conversion factor between a given base currency and a target currency
  """
  url = f'https://v6.exchangerate-api.com/v6/dea45ae89afd7434839c7abe/pair/{base_currency}/{target_currency}'

  response = requests.get(url)

  return response.json()

```

```python
@tool
def convert(base_currency_value: int, conversion_rate: Annotated[float, InjectedToolArg]) -> float:  #Now LLM doen't set it's value at its own
  """
  given a currency conversion rate this function calculates the target currency value from a given base currency value
  """

  return base_currency_value * conversion_rate
```

```python
get_conversion_factor.invoke({'base_currency':'USD','target_currency':'INR'})
```

```python
convert.invoke({'base_currency_value':10, 'conversion_rate':85.16})
```

```python
convert.args
```

```python
convert.invoke({'base_currency_value':100,'conversion_rate':91.8485})
```

```python
#tool binding:
llm=ChatGoogleGenerativeAI(model='gemini-3-flash-preview')
```

```python
llm_with_tools = llm.bind_tools([get_conversion_factor, convert])
```

```python
#tool calling
message = [HumanMessage('What is the conversion factor between INR and USD, and based on that can you convert 100 inr to usd and give me that result after conversion')]
```

```python
ai_message = llm_with_tools.invoke(message)
```
